```python
import numpy as np
import scipy.linalg
```

```python
A1 = np.array([[-9,  3, -7, -2, -4],
              [ 6, -6,  2,  4,  2],
              [-5,  3, -3, -2, -2]])
print(A1)
```

```python
# L should be 3x3 and U 3x5
```

```python
L1 = np.array([[   1,    0,   0],
              [-2/3,    1,   0],
              [ 5/9, -1/3,  1]])
print(L1)
```

```python
U1 = np.array([[-9,  3,   -7,  -2,   -4],
              [ 0, -4, -8/3, 8/3, -2/3],
              [ 0,  0,    0,   0,    0]])
print(U1)
```

```python
# check if our L and U are correct
A2 = np.dot(L1, U1)
print(A2); print(A1)
```

```python
# verify with the scipy LU decomposition
P, L, U = scipy.linalg.lu(A1)
# permutation matrix P is the identity in this example
L.shape, U.shape
```

```python
print(L); print(U)
```

```python
A3 = np.dot(L,U)
print(A3); print(A1)
```
