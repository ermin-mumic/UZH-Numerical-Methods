```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Some sample data points to interpolate
X = np.array([0, 1, 4, 9])
Y = np.array([0, 1, 2, 3])

# Dense sampling along the x-axis for reconstruction
newX = np.linspace(0, 9, 50)
```

```python
# Show the data points
plt.scatter(X, Y, color="orange", label='Sample points') # or use plot()
plt.legend()
plt.show()
```

Now define the basis functions to be used, e.g. powers of a polynomial.

```python
# Now define the basis functions to be used, e.g. polynomials of degree k.
def FuncXk(X,k):
    return X**k

plt.plot(newX, FuncXk(newX, 0), color="brown", label='Constant y = x')
plt.plot(newX, FuncXk(newX, 1), color="red",   label='Line y = x')
plt.plot(newX, FuncXk(newX, 2), color="green", label='Quadratic curve y = x^2')
plt.plot(newX, FuncXk(newX, 3), color="blue",  label='Cubic curve y = x^3')
plt.legend()
plt.show()
```

Want to find Ac = y, with A being the function values of 1, x, x^2, x^3. So need to fill A with these columns, and then solve for c = A-1 times y.

```python
A = np.array([FuncXk(X,0), FuncXk(X,1), FuncXk(X,2), FuncXk(X,3)]).T # transposed into column form
A.shape, A
```

```python
c = np.linalg.inv(A).dot(Y)
c
```

Ok, now we should test if the functions 1, x, 1, x, x^2, x^3 with coefficients c actually create the original curve. For this we compute the linear combination from the coefficients of c and the basis functions applied to the input values of X.

```python
def testFunc(c, newX):
    A = np.array([FuncXk(newX,0), FuncXk(newX,1), FuncXk(newX,2), FuncXk(newX,3)]).T
    return A.dot(c)
```

```python
X, testFunc(c,X)
```

Now apply the interpolation function (myFunc) to a new set of input values and plot it.

```python
newY = testFunc(c, newX)
plt.plot(X, Y, color="orange", label='Sample points')
plt.plot(newX, newY,           label='Interpolating cubic polynomial')
plt.legend()
plt.show()
```

Now let's check the rank of our system matrix A:

```python
np.linalg.matrix_rank(A)
```

Now use a different array of x input values with one x repeated:

```python
X = np.array([0, 1, 1, 9])
plt.scatter(X, Y, color="orange", label='Updated sample points')
plt.legend()
plt.show()
```

```python
B = np.array([FuncXk(X,0), FuncXk(X,1), FuncXk(X,2), FuncXk(X,3)]).T # transposed into column form
B
```

Now let's check this matrix' rank:

```python
np.linalg.matrix_rank(B)
```

```python
c = np.linalg.inv(B).dot(Y)
c
```

```python
newY = testFunc(c, newX)
plt.plot(X, Y, color="orange", label='Sample points')
plt.plot(newX, newY,           label='Interpolating cubic polynomial')
plt.legend()
plt.show()
```
